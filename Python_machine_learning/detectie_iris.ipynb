## Analiza datelor
</div>

Cu Colab, puteți valorifica biblioteci Python populare pentru a analiza și a vizualiza date. Celula de cod de mai jos folosește <strong>numpy</strong> pentru a genera date aleatorii și <strong>matplotlib</strong> pentru vizualizarea acestora. Pentru a edita codul, dați clic pe celulă și faceți modificări.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage.util import random_noise
from skimage import feature
from skimage.color import rgb2gray

from skimage.transform import hough_circle
from skimage import data, color
from skimage.feature import canny
# Import the necessary libraries
import numpy as np
from skimage.transform import hough_circle
from skimage.feature import canny
from skimage import data, color
import matplotlib.pyplot as plt
import skimage.io
import skimage.color
import numpy as np
```

```
uploaded = files.upload()
imgFileName=list(uploaded.keys())[0]
```

```
ImgIn = skimage.io.imread(imgFileName)
if len(ImgIn.shape) == 3:
  ImgGray = np.uint8(skimage.color.rgb2gray(ImgIn)*255)
else:
  ImgGray = ImgIn
```

```
fig, axes = plt.subplots(1, 1, figsize = (6,6))
axes.imshow(ImgIn, cmap='gray')
axes.axis('off')
axes.set_title('Imaginea de intrare')
plt.show()
```

```
#afisarea imaginii gri
fig, axes = plt.subplots(1, 1, figsize = (6,6))
axes.imshow(ImgGray, cmap='gray')
axes.axis('off')
axes.set_title('Imaginea de intrare pe nivele de gri')
plt.show()
```

```
#afisam histograma imaginii pe nivele de gri
fig, axes = plt.subplots(1, 2, figsize = (12,4))
axes[0].imshow(ImgGray, cmap='gray')
axes[0].axis('off')
axes[0].set_title('Luminanta imaginii de intrare')
axes[1].hist(ImgGray.ravel(), 256, [0,255], density = True)
axes[1].set_title('Histograma normalizata a luminantei imaginii de intrare')
plt.show()
```

```
# aplicăm filtrul Gaussian pentru reducerea zgomotului
# blurred = cv2.GaussianBlur(ImgGray, (9,9), 2)       # (75,75) = media valorilor pixelilor dintr o fereastra HxL
blurred = ndi.gaussian_filter(ImgGray, 15)                                                    # 0 = sigma (are efect asupra estomparii)
fig, axes = plt.subplots(1, 1, figsize = (6,6))
axes.imshow(blurred, cmap='gray')
axes.axis('off')
axes.set_title('Imaginea Blurată')
plt.show()
```

```
# aplicăm detectarea marginilor folosind detectorul Canny
edges = cv2.Canny(blurred, threshold1=9, threshold2=15)

fig, axes = plt.subplots(1, 1, figsize = (6,6))
axes.imshow(edges, cmap='gray')
axes.axis('off')
axes.set_title('Detecția Marginilor (Canny)')
plt.show()
```

```
 # Aplicăm transformarea Hough pentru detectarea cercurilor (irisului)
detected_circles = cv2.HoughCircles(
        edges,
        cv2.HOUGH_GRADIENT,
        dp=1.5,             # Factor de rezoluție al acumulării Hough
        minDist=100,        # Distanța minimă între centrele cercurilor detectate
        param1=50,
        param2=30,          # Prag pentru detectarea cercurilor
        minRadius=90,       # Setează raza minimă aproape de dimensiunea irisului
        maxRadius=200       # Setează raza maximă pentru a include doar irisul
    )
```

```
# Desenăm cercul cel mai mare detectat (presupunem că acesta este irisul)
def draw_iris_circle(ImgIn, detected_circles):
    if detected_circles is not None:
        detected_circles = np.uint16(np.around(detected_circles))

        # Alegem doar cercul cel mai mare
        max_radius = 0
        iris_circle = None
        for pt in detected_circles[0, :]:
            a, b, r = pt
            if r > max_radius:
                max_radius = r
                iris_circle = (a, b, r)

        # Dacă am găsit un cerc pentru iris, desenăm cercul pe imaginea originală
        if iris_circle is not None:
            a, b, r = iris_circle
            # Desenăm circumferința cercului
            ImgIn = cv2.circle(ImgIn, (a, b), r, (0, 255, 0), 2)
            # Desenăm centrul cercului
            ImgIn = cv2.circle(ImgIn, (a, b), 2, (0, 0, 255), 3)

    # Afișăm imaginea blurată și imaginea cu marginile detectate
    fig, axes = plt.subplots(1, 1, figsize=(6, 6))

    # Afișăm imaginea originală cu irisul încercuit
    plt.imshow(ImgIn)
    plt.title("Imaginea Originală cu Irisul Detectat")
    plt.axis('off')
    plt.show()

#call the draw iris function
draw_iris_circle(ImgIn, detected_circles)
```
